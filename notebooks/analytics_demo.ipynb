# Analytics Demo

Plots & EDA on extracted transaction data.


```python
import sys
sys.path.append('..')

import pandas as pd
import matplotlib.pyplot as plt
from app.db import SessionLocal
from app.models import Transaction
from app.services.insights import InsightsService
```

```python
# Load transactions
db = SessionLocal()
try:
    transactions = db.query(Transaction).all()
    df = pd.DataFrame([{
        'date': t.date,
        'amount': t.amount,
        'vendor': t.vendor,
        'category': t.category
    } for t in transactions])
    
    print(f"Total transactions: {len(df)}")
    print(df.head())
finally:
    db.close()
```

```python
# Plot spending over time
if not df.empty:
    df['date'] = pd.to_datetime(df['date'])
    df.set_index('date', inplace=True)
    df['amount'].resample('D').sum().plot()
    plt.title('Daily Spending')
    plt.ylabel('Amount')
    plt.show()
```
